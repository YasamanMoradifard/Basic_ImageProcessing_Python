# Basic Image Processing|

## 1. Required Packages and Data

```python
# Importing libraries
from scipy import misc, ndimage
from skimage import io
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Importing an image
## using skimage
#img = io.imread(image_url)

## using scipy
#img = misc.imread(image_url)

from skimage import data
img = data.retina()
```

```python
# check image data type
type(img)
```

```python
# check the shape and data type
img.shape, img.dtype
```

```python
# mean, max, min amount related to pixels
img.mean(), img.min(), img.max()
```

## 2. Image transformation

### 2.1. Flipping

```python
# Flip Left to Right
flipL2R = np.fliplr(img)

# Flip Upp to Down 
flipU2D = np.flipud(img)

# show result
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15,5))

ax0.set_title('original image')
ax0.imshow(img)

ax1.set_title('Left to Right')
ax1.imshow(flipL2R)

ax2.set_title('Up to Down')
ax2.imshow(flipU2D)
```

### 2.2. Rotation 

```python
img_rotated_90 = ndimage.rotate (img, 90)
img_rotated_180 = ndimage.rotate (img, 180)

# show result
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15,5))

ax0.set_title('original image')
ax0.imshow(img)

ax1.set_title('90 degree rotation')
ax1.imshow(img_rotated_90)

ax2.set_title('180 degree rotation')
ax2.imshow(img_rotated_180)
```

## 3. Filters

### 3.1. Uniform filter

```python
img = data.coins()
img_uniform_3 = ndimage.uniform_filter(img, size = 3)
img_uniform_7 = ndimage.uniform_filter(img, size = 7)

# show result
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15,5))

ax0.set_title('original image')
ax0.imshow(img)

ax1.set_title('filtered image, size=3')
ax1.imshow(img_uniform_3)

ax2.set_title('filtered image, size=7')
ax2.imshow(img_uniform_7)
```

### 3.2. Gaussian filter
> smooth the noise, but the edges are gone.

```python
img = data.coins()
img_gaussian_3 = ndimage.gaussian_filter(img, sigma = 3)
img_gaussian_7 = ndimage.gaussian_filter(img, sigma = 7)

# show result
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15,5))

ax0.set_title('original image')
ax0.imshow(img)

ax1.set_title('filtered image, sigma=3')
ax1.imshow(img_gaussian_3)

ax2.set_title('filtered image, sigma=7')
ax2.imshow(img_gaussian_7)
```

### 3.3. median filter
> When you want to keep edges as well.

```python
img = data.coins()
img_median_3 = ndimage.median_filter(img, 3)
img_median_7 = ndimage.median_filter(img, 7)

# show result
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15,5))

ax0.set_title('original image')
ax0.imshow(img)

ax1.set_title('filtered image, order=3')
ax1.imshow(img_median_3)

ax2.set_title('filtered image, order=7')
ax2.imshow(img_median_7)
```

### 3.4. Sobel filter
> filtering on just one of the image axis.

```python
img = data.cell()
img_sobel_1 = ndimage.sobel(img, axis = -1)
img_sobel_0 = ndimage.sobel(img, axis = 0)

# show result
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15,5))

ax0.set_title('original image')
ax0.imshow(img)

ax1.set_title('filtered image on axis=-1')
ax1.imshow(img_sobel_1)

ax2.set_title('filtered image on axis=0')
ax2.imshow(img_sobel_0)
```
